# Importando bibliotecas

```python
import numpy as np
import pandas as pd
from sympy import *
np.random.seed(100)
```

# Definindo o tamanho da amostra

```python
# Escolhemos uma amostra de 2000 pessoas para essa análise.
n = 2000
```

```python
"""Gerando média de altura de 1.75 e desvio padrão 0.30, 
sendo 2000 pessoas em uma distribuição normalizada."""
alt = np.random.normal(1.75, 0.30, n)
alt
```

```python
# Convertendo array para lista.
alt = alt.tolist()
```

# Gerando peso a partir da altura

```python
x1 = symbols('alt_i')
x2 = symbols('aleatorio_i')
peso_i = x1*50 + x2 #x2 varia no intervalo de -5 a 5 e segue distribuição uniforme.
peso_i
```

```python
peso = []
for i in alt:
    a = round(float(np.random.uniform(-5, 5)), 2)
    b = 50 * i
    c = a + b
    peso.append(c)
```

```python
# Verificando o tamanho da lista
len(peso)
```

```python
# Formula do IMC
peso1, altura1 = symbols(['peso', 'altura'])
IMC_formula = peso1 / (altura1**2)
IMC_formula
```

```python
IMC = []
for i, j in zip(peso,alt):
    e = i / (j**2)
    e = round(e, 2)
    IMC.append(e)
```

```python
IMC
```

```python
len(IMC)
```

# Gerando sexos

```python
sexo = []
for i in range(0,2000):
    value = np.random.uniform(0,1)
    if value > 0.5:
        sexo.append('M')
    else:
        sexo.append('F')
```

```python
sexo
```

```python
len(sexo)
```

# Gerando correspondencia de problema cardiaco

```python
cardiaco = []
for i in range(0,2000):
    if IMC[i] > 40:
        cardiaco.append('S')
    else:
        if IMC[i] > 30 and IMC[i] < 40:
            cardiaco.append('P')
        else:
            cardiaco.append('N')
```

```python
cardiaco
```

```python
data = alt,peso,IMC,sexo,cardiaco
```

```python
df = pd.DataFrame(data, index=['Altura', 'Peso', 'IMC', 'Sexo', 'Cardiaco']).T
```

# Frenquencia para valores qualitativos

```python
freq_m = 0
freq_f = 0
for i in df.Sexo:
    if i == 'M':
        freq_m = freq_m + 1
    else:
        freq_f = freq_f + 1 

print('O percentual de homens é: ', round((freq_m/2000*100),2))
print('O percentual de mulheres é: ', round((freq_f/2000*100),2))
```

```python
freq_s = 0
freq_p = 0
freq_n = 0

for i in df.Cardiaco:
    if i == 'S':
        freq_s = freq_s + 1
    elif i == 'P':
        freq_p = freq_p + 1
    else:
        freq_n = freq_n + 1

print('O percentual de pessoas com problemas cardíacos é: ', round((freq_s/2000*100),2))
print('O percentual de pessoas com tendência a problemas cardíacos é: ', round((freq_p/2000*100),2))
print('O percentual de pessoas sem tendência a problemas cardíacos é: ', round((freq_n/2000*100),2))
```
